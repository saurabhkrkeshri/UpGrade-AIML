**Question 3:**
Write a Python program to check if the given string is a Palindrome or Anagram or None of them. Display
the message accordingly to the user.

```
x = input("Enter anything")
y = input("Enter another thing")
l1 = list(x)
l2 = list(y)
temp = l1
k1 = l1
k2 = l2
k1 = k1.sort()
k2 = k2.sort()
temp.sort()
temp.reverse()
if temp == l1:
  print("Palindrome")
if k1 == k2:
  print("Anagram")
if k1 != k2 and temp != l1:
  print("Nothing")
```
