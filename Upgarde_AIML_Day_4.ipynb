**Question 2 :** Research on range() functions and its parameters. Create a markdown cell and write in your own words (no copy-paste from google please) what you understand about it. Implement a small program of your choice on the same. 

**Solution:** range() fuction is a sequence constructor. It creates a sequence of whole numbers.

> 

*   Range function can take 3 arguments
*   Syntax: range(start,end,step)


*   "start" and "step parameters are optional arguments
*   Use of range() with "end" argument. e.g. range(5) --> 0, 1, 2, 3, 4


*   Here we can make a observation that default starting value is 0 and default step value is 1. So, it starts with 0 and increments of 1 that's why we get the sequence 0, 1, 2, 3, 4
*   One more important observation is that ending value of sequence must be lesser than "end" value.


*   Now, we see all the optional arguments:
range(10,23,6) ---> 10, 16, 21
*   Clearly, we can see starting value of 10 with increment of 6 till before than "end"(here 23) value.

Range function can be used to create reverse sequence also.


*   For this we can set value of "start" > "end" with negative stepping.
*   e.g. range(10,5,-2) --> 10, 8, 6



# *   "for" loop in python is the best method to access the elements of range()






















```
for e in range(36,2,-5):
  print(e)
```

**Question 4:** Consider a list of 10 elements of integer values. If the number in the list is divisible by 2, print the result as "square of that number minus 2". 

```
x = eval(input("Enter the list elements"))
for e in x:
  if e%2==0:
    print(e**2-2)
```
